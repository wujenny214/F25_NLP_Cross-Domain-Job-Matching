```python
import pandas as pd

## Raw dataset
data= pd.read_csv("../0_raw_dataset/Resume.csv")
```

```python
data.head()
```

```python
resume_df=data[[
    "ID",
    "Resume_str", "Category"
]]
resume_df
```

```python
exclude = ["CHEF", "AGRICULTURE", "BPO"]

resume_df_s = (
    resume_df[~resume_df["Category"].isin(exclude)]
    .sample(n=100, random_state=42)
)
resume_df_sample=resume_df_s[[
    "ID",
    "Resume_str"
]]
resume_df_sample.to_csv("../1_data_cleaning/resume_cleaned.csv", index=False)
resume_df_s.to_csv("../5_checking_accuracy/Resume.csv", index=False)
```
