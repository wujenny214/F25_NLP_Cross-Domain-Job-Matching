## Job Description Dataset Embedding

```python
import pandas as pd
```

```python
jd_df = pd.read_excel("../1_data_cleaning/filtered_jd_sections2.xlsx")
```

```python
jd_df.head()
```

```python
# create semantic sentence for each row
def build_semantic_sentence(row):
    desc = str(row["job_description"])
    loc  = str(row["location_cleaned"])
    title = str(row["job_title"])
    duties = str(row["jd_duties"])
    req = str(row["jd_requirements"])
    edu = str(row["jd_education"])

    return (
        f"This job, titled {title} and located in {loc}, involves {duties}. "
        f"The role description states: {desc}. "
        f"Applicants are expected to meet the following requirements: {req}. "
        f"The position lists the following educational background: {edu}."
    )

jd_df["semantic_sentence"] = jd_df.apply(build_semantic_sentence, axis=1)
jd_df.head()

```

```python
from sentence_transformers import SentenceTransformer
import torch

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
def embed_sentence(sentence):
    return model.encode(sentence)
jd_df["semantic_emb"] = jd_df["semantic_sentence"].apply(embed_sentence)
jd_df.head()
```
